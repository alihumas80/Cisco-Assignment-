```python
#Question 5

class Cars():
    def __init__(self,name,model,color,weight,seats):
        self.name = name
        self.model = model
        self.color = color
        self.weight = weight
        self.seats = seats
    def update_car_model(self,newmodel):
        self.updatemodel = newmodel
    def update_car_color(self,newcolor):
        self.updatecolor = newcolor

Car1 = Cars('Audi','A1','Black','1882kg','5')
print(Car1.name)
print(Car1.model)
print(Car1.color)
print(Car1.weight)
print(Car1.seats)
Car1.update_car_model('A2')
print(Car1.updatemodel,' NEW Model \n')

Car2 = Cars('Bentley','Bentley T-Series','White','1885kg','6')
print(Car2.name)
print(Car2.model)
print(Car2.color)
print(Car2.weight)
print(Car2.seats)
Car2.update_car_color('Green')
print(Car2.updatecolor,' NEW Color \n')

Car3 = Cars('BMW','BMW 640','Red','1666kg','5')
print(Car3.name)
print(Car3.model)
print(Car3.color)
print(Car3.weight)
print(Car3.seats)
Car3.update_car_model('BMW 740')
print(Car3.updatemodel,' NEW Model \n')

Car4 = Cars('Bugatti','Veyron 16.4','Blue','1888kg','5')
print(Car4.name)
print(Car4.model)
print(Car4.color)
print(Car4.weight)
print(Car4.seats)
Car4.update_car_color('Gray')
print(Car4.updatecolor,' NEW Color\n')

Car5 = Cars('Buick','Buick Lacrosse','Brown','1999kg','6')
print(Car5.name)
print(Car5.model)
print(Car5.color)
print(Car5.weight)
print(Car5.seats)
Car5.update_car_model('Buick 123')
print(Car5.updatemodel,' NEW Model ')
```
